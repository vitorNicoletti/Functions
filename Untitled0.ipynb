```python
import matplotlib.pyplot as plt
import numpy as np

r = 1
ang = 45
theta = np.linspace(0, 2*np.pi, 100)
x = r * np.cos(theta)
y = r * np.sin(theta)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, y)
ax.set_title('Círculo no Plano Cartesiano')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.grid(True)
ax.axis('equal')

# Criando um novo eixo para a linha
seno = [0,np.sin(np.deg2rad(ang))]
cos = [0,np.cos(np.deg2rad(ang))]
tang = [0,np.tan(ang)]
print(cos,seno,tang)
ax.plot([0,0],seno)
ax.plot(cos,[0,0])
ax.plot()
ax.plot(cos, seno, label="Linha")

plt.show()
```
